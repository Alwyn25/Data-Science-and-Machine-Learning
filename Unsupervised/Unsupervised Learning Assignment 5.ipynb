```python
# Importing pandas
import pandas as pd
```

```python
# Using pandas import the dataset as dataframe
dataframe=pd.read_csv('Retail.csv',encoding='unicode_escape',header=None)
dataframe.head()
```

```python
dataframe.shape
```

```python
 dataframe = dataframe.drop(dataframe.columns[[1,5]],axis=1)
```

```python
dataframe.head(19)
```

```python
new_dataframe = dataframe.iloc[2:11,:]
new_dataframe.head(9)
```

```python
new_dataframe.shape
```

```python
new_dataframe.columns=['Transactions from the Database','Item','No: of transactions in which the item is bought','Support']
```

```python
new_dataframe.head()
```

```python
list_data_frame=[]
for i in range(0,9):
    list_data_frame.append([str(new_dataframe.values[i,j]) for j in range (0,4)])
```

```python
list_data_frame
```

```python
# Install the mixtend library to use apriory and association rule mining
from apyori import apriori
```

```python
association_rules=apriori(list_data_frame,min_support=0.05,min_confidence=0.7,min_lift=1.2,min_length=2)
association_result=list(association_rules)
```

```python
# Using the apriori algorithm generate a list of item frequently brought together.
# Generate the association rules for the given items from apriori algorithm
association_result
```
