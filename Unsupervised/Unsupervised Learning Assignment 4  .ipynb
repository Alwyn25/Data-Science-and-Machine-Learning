```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy as sch
```

```python
#Using the np.array function create an np array from the given data.
data = np.array([[5,3],[10,15],[15,12],[24,10],[30,30],[85,70],[71,80],[60,78],[70,55],[80,91]])
# Generating scatter plot for the data.
plt.scatter(data[:,0],data[:,1],edgecolors='black', color= 'red')
```

```python
# Ploting dendrogram to visualize the clustering linkage
clustering_linkage = sch.linkage(data)
sch.dendrogram(clustering_linkage)
plt.show()
```
